```python
cinema = pd.read_csv('https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv')
cinema
```

Dataframe 'Cinema' имеет 77 строк и 8 колонок: 'Film', 'Genre', 'Lead Studio', 'Audience score %', 'Profitability',
       'Rotten Tomatoes %', 'Worldwide Gross', Dataframe 'Cinema'не имеет пустых значений, имеет дублика 
1.Film - колонка типу object, вмещает в себе названия фильмов.
 - количество фильмов  - 75 (были удалены дублякаты для: "Mamma Mia!",	"Gnomeo and Juliet"  )
2.Genre - колонка типу object , информирует к какому жанру пренадлежит наша кинокартина
 - представленно 6 жанров: 'Romance', 'Comedy', 'Drama', 'Animation', 'Fantasy', 'Action'
 - болше всего фильмов жанра  "Comedy" - 42, меньше всего "Fantasy" -1,"Action"-1.
 - 
3.Lead Studio - колонка типу object, информирует о компании, которая сняла фильм
4.Audience score % - колонка типу int64,выражается в процентрах,так как представляет собой процент пользователей, положительно оценивших фильм.
5.Profitability - колонка типу float64, отображает прибыльность картины, которая расчитывается по формуле прибыльность = сборы / бюджет.
6.Rotten Tomatoes % - колонка типу int64, выражается в процентрах, данные пердосталяет сайт-агрегатор рецензий, который подсчитывает процент положительных отзывов и средний балл в рецензиях, опубликованных другими изданиями.
7.Worldwide Gross - колонка типу object,обозначает мировой брутто, что выражается в сумме внутреного дохода и международного дохода от картины.
8.Year - колонка типу int64,показывает год выхода картины.

# После подсчета количества фильмов,было определено, что два фильма имеют дубликаты с аналогичными значениями.Данные дубликаты быдут выдалены для избежания влиянния на подсчет статистики.

```python
#Удаляем дубликаты 
del_dup_cinema = cinema.drop_duplicates(keep='last')
del_dup_cinema
```

```python
#Проверяем операцию по удалению дубликатов
del_dup_cinema['Film'].value_counts()
```

```python
#Выводим уникальные значение по колонке 'Genre'
cinema['Genre'].unique()
```

```python
#Выводим количестово каждого значение в колонке 'Genre'
del_dup_cinema['Genre'].value_counts()
```

# В ходе просмотра данных по колонке "Genre" было установаленно, что названия  имеют дублика в нижнем регистре, что повлияет на дальнейшую работу с данными.

```python
#Редактируем некоректные значения для по колонки 'Genre'
del_dup_cinema['Genre'] = del_dup_cinema['Genre'].replace(['romance','Romence'],'Romance')
```

```python
#Проверяем значение 'Comdy' для по колонки 'Genre'
del_dup_cinema[del_dup_cinema['Genre'] == 'Comdy']
```

# Для уточнения информации по  'Comdy' для 'Genre' , произведен поиск по названию фильма.

```python
#Редактируем некоректные значения для по колонки 'Genre'
del_dup_cinema['Genre'] = del_dup_cinema['Genre'].replace(['comedy','Comdy'],'Comedy')
```

```python
#Выводим количестово каждого значение в колонке 'Genre' после корретировки 
del_dup_cinema['Genre'].value_counts()
```

```python
#Проверяем уникальные значение в колонке 'Genre' после корретировки
del_dup_cinema['Genre'].unique()
```

```python
#Удаляем символ "$" для дальнейшей работі с колонкой 'Worldwide Gross' и изменяем тип для проведение математических операций
del_dup_cinema['Worldwide Gross']= del_dup_cinema['Worldwide Gross'].str.replace('$', '')
del_dup_cinema[['Worldwide Gross']].astype(float)
```
