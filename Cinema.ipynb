>После подсчета количества фильмов,было определено, что два фильма имеют дубликаты с аналогичными значениями.Данные дубликаты быдут выдалены для избежания влиянния на подсчет статистики.

```python
#Удаляем дубликаты 
del_dup_cinema = cinema.drop_duplicates(keep='last')
del_dup_cinema
```

```python
#Проверяем операцию по удалению дубликатов
del_dup_cinema['Film'].value_counts()
```

```python
#Выводим уникальные значение по колонке 'Genre'
cinema['Genre'].unique()
```

```python
#Выводим количестово каждого значение в колонке 'Genre'
del_dup_cinema['Genre'].value_counts()
```

> В ходе просмотра данных по колонке "Genre" было установаленно, что названия  имеют дубликаты с некорректными названиями, что повлияет на дальнейшую работу с данными.

```python
#Редактируем некоректные значения для по колонки 'Genre'
del_dup_cinema['Genre'] = del_dup_cinema['Genre'].replace(['romance','Romence'],'Romance')
```

```python
#Редактируем некоректные значения для по колонки 'Genre'
del_dup_cinema['Genre'] = del_dup_cinema['Genre'].replace(['comedy','Comdy'],'Comedy')
```

```python
#Выводим количестово каждого значение в колонке 'Genre' после корретировки 
del_dup_cinema['Genre'].value_counts()
```

```python
#Проверяем уникальные значение в колонке 'Genre' после корретировки
del_dup_cinema['Genre'].unique()
```

```python
#Удаляем символ "$" для дальнейшей работы с колонкой 'Worldwide Gross'
del_dup_cinema['Worldwide Gross']= del_dup_cinema['Worldwide Gross'].str.replace('$', '')
```

```python
#Берем колонку 'Worldwide Gross' и преобразовуем в числовой для подальшого переобразования в тип float 
del_dup_cinema['Worldwide Gross'] = pd.to_numeric(del_dup_cinema['Worldwide Gross'],errors = 'coerce')
```

```python
#Переобразования в тип float
del_dup_cinema['Worldwide Gross'].astype(float)
```

```python
#Проверяю тип колонки после переобразования 
del_dup_cinema['Worldwide Gross'].dtype
```

```python
#Спомошью групировки по колонке 'Genre' получаем максимальные значения для каждого жанра
del_dup_cinema.groupby('Genre').agg({
  "Audience score %": "max",
  "Profitability": "max",
  "Rotten Tomatoes %": "max",
  "Worldwide Gross": "max"  
}).reset_index()
```

```python
#Спомошью групировки по колонке 'Genre' получаем минимальное значения для каждого жанра 
del_dup_cinema.groupby('Genre').agg({
  "Audience score %": "min",
  "Profitability": "min",
  "Rotten Tomatoes %": "min",
  "Worldwide Gross": "min"  
}).reset_index()
```

```python
#Спомошью группировки по колонке 'Genre' вычисляем среднее значения для каждого жанра  
del_dup_cinema.groupby('Genre').agg('mean')
```

```python
#Выволим количество фильмов для каждой студии 
del_dup_cinema['Lead Studio'].value_counts()
```

```python
#С помошью групировки по колонке 'Lead Studio' получаем максимальные значения для каждой Lead Studio 
del_dup_cinema.groupby('Lead Studio').agg({
  "Audience score %": "max",
  "Profitability": "max",
  "Rotten Tomatoes %": "max",
  "Worldwide Gross": "max"  
}).reset_index()
```

```python
#С помошью групировки по колонке 'Lead Studio' получаем минимальное значения для каждой Lead Studio
del_dup_cinema.groupby('Lead Studio').agg({
  "Audience score %": "min",
  "Profitability": "min",
  "Rotten Tomatoes %": "min",
  "Worldwide Gross": "min"  
}).reset_index()
```

```python
#С помошью группировки по колонке 'Lead Studio' вычисляем среднее значения для каждой Lead Studio  
del_dup_cinema.groupby('Lead Studio').agg('mean')
```

```python
#Выводим самые прибыльные фильмы
del_dup_cinema.sort_values(by='Profitability', ascending=False).head()
```

```python
#Выводим самые убыточные фильмы 
del_dup_cinema.sort_values(by='Profitability', ascending=False).tail()
```

```python
#Подробную информацию о наименее и наиболее прибыльных фильмах в наборе данных
del_dup_cinema.loc[del_dup_cinema['Film'].isin(['Fireproof','Jane Eyre','Our Family Wedding','When in Rome'])].sort_values(by='Profitability', ascending=False)
```

```python
#Сохраняем 10 лучших комедий по просмотру 
top_comedy=del_dup_cinema[del_dup_cinema['Genre'] == 'Comedy'].sort_values(by='Audience score %', ascending=False).head(10)[['Film','Year','Lead Studio']]
top_comedy.to_csv('top_comedy.csv')
```
