```python
#Преобразовуем в числовой тип
powerlift['best3squat_kg'] = pd.to_numeric(powerlift['best3squat_kg'],errors = 'coerce')
```

```python
#Преобразовуем в числовой тип
powerlift['best3deadlift_kg'] = pd.to_numeric(powerlift['best3deadlift_kg'],errors = 'coerce')
```

```python
#Преобразовуем в числовой тип
powerlift['best3bench_kg'] = pd.to_numeric(powerlift['best3bench_kg'],errors = 'coerce')
```

```python
#Проверяем изменения типа
powerlift.info('Dtype')
```

```python
#Рекорды каждого пола
powerlift.groupby('sex').agg({
  "best3squat_kg": "max",
  "best3bench_kg": "max",
  "best3deadlift_kg": "max"
}).reset_index()
```

```python
#Рекорды каждого дивизиона
powerlift.groupby('division').agg({
  "best3squat_kg": "max",
  "best3bench_kg": "max",
  "best3deadlift_kg": "max"
}).reset_index()
```

```python
#Промотр уникальных 
powerlift['place'].unique()
```

```python
#Заменяем нечисловые значения
powerlift['place'] = powerlift['place'].replace(['DD','DQ','G'],'0')
```

```python
#Возвращает новый DataFrame, в котором типы данных были изменены на тип int
powerlift['place'] = powerlift['place'].astype('int')
```

```python
#Выводим в новый DataFrame участников занявшие первые места
powerlift1 = powerlift[powerlift['place'] == 1]
powerlift1
```

```python
#С помощью pivot table суммируем победы для каждого учасника
powerlift_first_p = pd.pivot_table(
    powerlift1,index = 'name',
    values = 'place',
    aggfunc = sum)
powerlift_first_p.sort_values(by = 'place', ascending = False)
```

```python
#Сохраняем в csv файл
powerlift_first_p.to_csv('powerlift_first_p.csv')
```

```python
#С помощью pivot table выводим для каждой комбинации подразделения и пола участника с наибольшим количеством побед 
#среди участников
pd.pivot_table(powerlift1,
               index=['division','sex','name'],
               values='place',
               aggfunc=np.sum)
```
