```python
#Будем считать выбросами все точки, выходящие за пределы трёх сигм
def outliers_indices(feature):
    mid = coffee_df[feature].mean()
    sigma = coffee_df[feature].std()
    return coffee_df[(coffee_df[feature] < mid - 3*sigma) | (coffee_df[feature] > mid + 3*sigma)].index
```

```python
#Создаем переменный для выбросов по каждому показателю и выводим суммарное количество выбросов
aftertaste2 = outliers_indices('aftertaste')
balance2 = outliers_indices('balance')
body2 = outliers_indices('body')
acidity2 = outliers_indices('acidity')
sweetness2 = outliers_indices('sweetness')
flavor2 = outliers_indices('flavor')
clean_cup2 = outliers_indices('clean_cup')
aroma2 = outliers_indices('aroma')
uniformity2 = outliers_indices('uniformity')
cupper_points2 = outliers_indices('cupper_points')

out = set(aftertaste2) | set(balance2) | set(body2) | set(acidity2) | set(sweetness2) | set(flavor2) | set(clean_cup2) | set(aroma2) | set(uniformity2) | set(cupper_points2)

print(len(out))
```

><h4>Из набора данных будет удалено 114 объектов-выбросов, что не является существенным в данном случае</h4>

```python
#Удаляем выбросы 
coffee_df.drop(out, inplace=True)
```

><h4>3.Данные не репрезентативны для получнения зависимоть между цветом кофейных зерен и сортом кофе</h4>

```python
#Минимальные значения total_cup_points по странам экспортерам 
coffee_df.groupby('country_of_origin').agg({
  "total_cup_points": "min"
}).reset_index().sort_values(by='total_cup_points', ascending=True).head(20)
```

```python
#Средние значения total_cup_points по странам экспортерам
coffee_df.groupby('country_of_origin').agg({
  "total_cup_points": "mean"
}).reset_index().sort_values(by='total_cup_points', ascending=False).head(20)
```

```python
#Максимальные значения total_cup_points по странам экспортерам
coffee_df.groupby('country_of_origin').agg({
  "total_cup_points": "max"
}).reset_index().sort_values(by='total_cup_points', ascending=False).head(20)
```

```python
#C помощью seaborn.subplots визуализируем total_cup_points по country_of_origin
#для анализа зависимости между страной производителем кофе и качестом кофейных зерен
plt.subplots(figsize = (14,10))
sns.barplot(x='total_cup_points', y='country_of_origin',data=coffee_df)
plt.xlim([72, 87])
```

><h4>4.Анализируя данные по country_of_origin можно сказать, что страна происхождения кофе влияет на качество кофе. 
Лидером качества является - Ethiopia</h4>

```python
#C помощью seaborn.jointplot визуализируем total_cup_points по country_of_origin
#для анализа зависимости между страной высотой выращивания кофе и качестом кофейных зерен
sns.set_theme(style="darkgrid")

sns.jointplot(x ='altitude_mean_meters', y='total_cup_points',data=coffee_df,
                  kind ="reg", truncate=False,
                  xlim=(0,2500 ), ylim=(70, 90),
                  color="m", height=7)
```

```python
#C помощью seaborn.relplot визуализируем total_cup_points по country_of_origin
sns.relplot(x='total_cup_points', 
            y='altitude_mean_meters',
            data=coffee_df,
            kind='line')
plt.ylim([750, 2300])
```

><h4>5.Визуализация total_cup_points по country_of_origin показала, что существует зависимость между высотой выращивания кофейных зерен и качаством напитка</h4>
